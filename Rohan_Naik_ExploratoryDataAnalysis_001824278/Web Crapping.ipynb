```python
def GetValidLinks(u):
    try:
        i=dt.now()
        res=r.head(u)
        if u.startswith('http'):
            
            if res.status_code == 200:
                return (True,i.strftime('%Y/%m/%d %H:%M:%S'))
            else:
                return (False,i.strftime('%Y/%m/%d %H:%M:%S'))
        else:
            return (False,i.strftime('%Y/%m/%d %H:%M:%S'))
    except Exception:
        return (False,i.strftime('%Y/%m/%d %H:%M:%S'))
```

```python
GetValidLinks('https://www.investopedia.com/')
```

```python
url='https://www.investopedia.com/'
```

```python
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
col=['Link Name','Link url','IsValid','lastCheck']
dfr=pd.DataFrame(columns=col)
```

```python
source_code=r.get('https://www.investopedia.com/')
plain_text=source_code.text
sp=BeautifulSoup(plain_text,'html.parser')
o=10
l=0
dob=dt.now()
```

```python
for link in sp.find_all('a'):
    op=GetValidLinks(link.get('href'))
    x=[link.string,link.get('href'),op[0],op[1]]
    
    dfr.loc[l]=x
    l=l+1
```

```python
dfr.to_csv('WebScrapperEx.csv',index=False)
```
